# 백준 제출용

```python
li = [0,0,1,1]
x = int(input())

def cla(n):
    
    th = 500000
    tw = 500000
    
    if n%3 == 0:
        th = li[int(n/3)]
    if n%2 == 0:
        tw = li[int(n/2)]

    li.append(min([th,tw,li[n-1]]) + 1)

for i in range(4,x+1):
    cla(i)
    
print(li[x])
```
